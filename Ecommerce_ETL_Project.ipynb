#### Create database and create tables in the database

```python
cur, conn = create_database()
```

```python
#create the table for AccountData
AccountData_table_create = ("""CREATE TABLE accounts_data (
country_name VARCHAR,
country_code VARCHAR,
indicator_name VARCHAR,
series_code VARCHAR,
year_2015 numeric,
year_2016 numeric,
year_2017 numeric,
year_2018 numeric)
""")
cur.execute(AccountData_table_create)
conn.commit()
```

```python
#create the table for AccountCountry
AccountCountry_table_create = ("""CREATE TABLE account_country (
country_code VARCHAR PRIMARY KEY,
long_name VARCHAR,
table_name VARCHAR,
short_name VARCHAR,
currency_unit VARCHAR)
""")
cur.execute(AccountCountry_table_create)
conn.commit()
```

```python
#create the table for AccountSeries
AccountSeries_table_create = ("""CREATE TABLE account_series (
series_code VARCHAR PRIMARY KEY,
indicator_name VARCHAR,
long_definition VARCHAR,
topic VARCHAR)
""")
cur.execute(AccountSeries_table_create)
conn.commit()
```
